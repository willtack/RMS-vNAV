```python
import scipy.io
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", font_scale=1.3)
accumbens_dist = scipy.io.loadmat('../distributions/accumbens.mat')['D']
caudate_dist = scipy.io.loadmat('../distributions/caudate.mat')['D']
pallidum_dist = scipy.io.loadmat('../distributions/pallidum.mat')['D']
putamen_dist = scipy.io.loadmat('../distributions/putamen.mat')['D']
```

```python
# Whole figure
f,axes = plt.subplots(2,2, figsize=(20, 16), sharey=True)
a1 = sns.distplot(accumbens_dist, kde=False, ax=axes[0,0], color='r', hist_kws={'alpha': 0.75});
a2 = sns.distplot(caudate_dist, kde=False, ax=axes[0,1], color='r', hist_kws={'alpha': 0.75});
a3 = sns.distplot(pallidum_dist, kde=False, ax=axes[1,0], color='r', hist_kws={'alpha': 0.75});
a4 = sns.distplot(putamen_dist, kde=False, ax=axes[1,1], color='r', hist_kws={'alpha': 0.75});

# Set common labels
f.text(0.5, 0.04, 'Diff Score', ha='center', va='center')
f.text(0.06, 0.5, 'Raw Count', ha='center', va='center', rotation='vertical')
plt.suptitle('Distribution of Diff Scores')

# Axes-specific labels
a1.set_xlabel("Nucleus Accumbens")
a2.set_xlabel("Caudate")
a3.set_xlabel("Pallidum")
a4.set_xlabel("Putamen")

plt.savefig('3_nice/distributions.svg')
plt.savefig('3_nice/distributions.png')
plt.show()
```

```python
def get_bin_number(dist, observed, bins):
    N, bins, patches = plt.hist(dist, bins)
    plt.close()
    #loop through bins, find which bin observed value is in, get index
    i=0
    for i in range(0, len(bins)):
        if observed < bins[i]:
            return i-1
        i=i+1
```

```python
# Global values
color1 = sns.xkcd_palette(['denim blue'])
color2 = 'xkcd:faded red'
ylim=500
bins=50
figsize=(12, 8)
```

```python
# Individual Figures - Accumbens
acc_obvs = 0.0097
acc_i = get_bin_number(accumbens_dist, acc_obvs, bins)
plt.figure(figsize=figsize)
ax = sns.distplot(accumbens_dist, bins=bins, color=color1, norm_hist=False, hist_kws={'alpha': .95}, kde=False)
ax.patches[acc_i].set_facecolor(color2)
ax.set_xlabel("Diff Score")
ax.set_ylabel("Raw Count")
ax.set_ylim(0, ylim)
ax.set_title(r'Distribution of Diff Scores: Nucleus Accumbens')
plt.savefig('3_nice/accumbens.svg')
plt.savefig('3_nice/accumbens.png')
plt.show()
```

```python
# Individual Figures - Caudate
cau_obvs = 0.0153
cau_i = get_bin_number(caudate_dist, cau_obvs, bins)
plt.figure(figsize=figsize)
ax = sns.distplot(caudate_dist, bins=bins, color=color1, norm_hist=False, hist_kws={'alpha': 0.95}, kde=False)
ax.patches[cau_i].set_facecolor(color2)
ax.set_xlabel("Diff Score")
ax.set_ylabel("Raw Count")
ax.set_ylim(0, ylim)
ax.set_title(r'Distribution of Diff Scores: Caudate')
plt.savefig('3_nice/caudate.svg')
plt.savefig('3_nice/caudate.png')
plt.show()
```
